## DATA COLLECTION AND PROCESSING
#### IMPORT NECESSARY LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
```

```python
data = pd.read_csv("Heart Disease data.csv")

# Show sample data
data.head()
```

```python
# Show the short information of the dataset
data.info()
```

Total entry/row in the dataset are 1025 and no missing values and all the columns are numeric. Data is clean enough. There are 14 columns:
* age: The person's age in years
* sex: The person's sex (1 = male, 0 = female)
* cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)
* trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)
* chol: The person's cholesterol measurement in mg/dl
* fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)
* restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)
* thalach: The person's maximum heart rate achieved
* exang: Exercise induced angina (1 = yes; 0 = no)
* oldpeak: ST depression induced by exercise relative to rest
* slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)
* ca: The number of major vessels (0-3)
* thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)
* target: Heart disease (0 = no, 1 = yes)



```python
#Checking null values
data.isnull().sum()
```

## ANALYZING THE DATA
#### FINDING HEART DESEASE

```python
target=data.groupby('target').size()
print(target, '\n')
print('People having heart disease:', target.loc[1])
print('People dont have heart disease:', target.loc[0])
```

```python
#Converting Numerical Data into Categorical Data

def heart_disease(row):
    if row==0:
        return 'Absence'
    elif row==1:
        return 'Presence'
```

```python
data['Heart_Disease']=data['target'].apply(heart_disease)
data.head()
```

```python
hd=data.groupby('Heart_Disease')['target'].count()
hd
```

#### VISUALIZE DATA

```python
#Pie Chart Creation of Heart Disease Population % using MatplotLib

plt.figure(figsize=(5,5))
plt.pie(hd, labels=['Absence','Presence'], autopct='%0.0f%%')
plt.title('Heart Disease Population %', fontsize=20)
plt.show()
```

```python
#Countplot Creation of Population Age using MatplotLib and Seaborn

plt.figure(figsize=(15,7))
sns.countplot(x='age', data=data)
plt.title('Population Age', fontsize=17)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
#Statistical Analysis

Min_Age=data['age'].min()
Max_Age=data['age'].max()
Mean_Age=data['age'].mean()
print("Minimum Age =",Min_Age)
print("Maximum Age =",Max_Age)
print("Mean Age =",Mean_Age)
```

```python
#Categorical Analysis

Young_Ages=data[(data['age']>=29) & (data['age']<40)]
Middle_Ages=data[(data['age']>=40) & (data['age']<55)]
Elderly_Ages=data[(data['age']>55)]
print('Young Ages =',len(Young_Ages))
print('Middle Ages =',len(Middle_Ages))
print('Elderly Ages =',len(Elderly_Ages))
```

```python
#Bar Plot Creation of Age Category using MatplotLib and Seaborn

sns.barplot(x=['Young_Ages','Middle_Ages','Elderly_Ages'], y=[len(Young_Ages), len(Middle_Ages), len(Elderly_Ages)])
plt.title('Age Category', fontsize=17)
plt.xlabel('Age Range', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
#Converting Numerical Data into Categorical Data

def gender(row):
    if row==1:
        return 'Male'
    elif row==0:
        return 'Female'
```

```python
#Applying converted data into our dataset with new column - sex1

data['sex1']=data['sex'].apply(gender)
data.head()
```

```python
#Converting Numerical Data into Categorical Data

def age_range(row):
    if row>=29 and row<40:
        return 'Young Age'
    elif row>=40 and row<55:
        return 'Middle Age'
    elif row>55:
        return 'Elder Age'
```

```python
#Applying converted data into our dataset with new column - Age_Range

data['Age_Range']=data['age'].apply(age_range)
data.head()
```

```python
#Swarm Plot Creation of Gender Based Age Category using MatplotLib and Seaborn

plt.figure(figsize=(20,10))
sns.swarmplot(x='Age_Range', y='age', hue='sex1', data=data, order=['Young Age','Middle Age','Elder Age'], palette=['#ffc300', '#003566'], size=5)
plt.title('Gender Based Age Category', fontsize=17)
plt.xlabel('Age Category', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 11))
axes = axes.flatten()

# Count Plot Creation of Heart Disease Based On Age Category in first subplot
hue_order=['Young Age', 'Middle Age', 'Elder Age']
sns.countplot(x='Heart_Disease', hue='Age_Range', data=data, ax=axes[0], order=['Presence','Absence'], hue_order=hue_order, palette='twilight_shifted')
axes[0].set_title('Heart Disease by Age Range', fontsize=15, fontweight='bold')

#Count Plot Creation of Heart Disease Based on Gender in second subplot
sns.countplot(x=data['Heart_Disease'], hue='sex1', data=data, palette='BuGn_r', ax=axes[1])
axes[1].set_title('Heart Disease Based on Gender', fontsize=15, fontweight='bold')

#Count Plot Creation of Chest Pain Experienced in 3rd subplot
sns.countplot(x=data['Heart_Disease'], hue='cp', data=data, order=['Presence','Absence'], ax=axes[2])
axes[2].set_title('Chest Pain Experienced', fontsize=15, fontweight='bold')
handles, labels = axes[2].get_legend_handles_labels()
axes[2].legend(handles, ['Typical Angina', 'Atypical Angina', 'Non-Anginal pain', 'Asymptomatic'], title='Chest Pain Type')

#Count Plot Creation of Chest Pain Based On Gender in 4th plot
sns.countplot(x=data['sex1'], hue='cp', data=data, palette='twilight_shifted', ax=axes[3])
axes[3].set_title('Chest Pain Based On Gender', fontsize=15, fontweight='bold')
handles, labels = axes[2].get_legend_handles_labels()
axes[3].legend(handles, ['Typical Angina', 'Atypical Angina', 'Non-Anginal pain', 'Asymptomatic'], title='Chest Pain Type')


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 11))
axes = axes.flatten()

# Bar Plot Creation of Person's Resting Blood Pressure (mm Hg) in first subplot
sns.barplot(x='sex1', y='trestbps', data=data, ax=axes[0],hue='sex1', legend=False,  palette='twilight_shifted')
axes[0].set_title('Blood Pressure', fontsize=15, fontweight='bold')

# Bar Plot Creation of Cholestrol Level Based On Gender in second subplot
sns.barplot(x='sex1', y='chol', data=data, ax=axes[1], hue='sex1', legend=False,  palette='twilight_r')
axes[1].set_title('Cholestrol Level Based On Gender', fontsize=15, fontweight='bold')

# Line Plot Creation of Blood Pressure VS Age in 3rd subplot
sns.lineplot(x='age', y='trestbps', data=data, color='r', ax=axes[2])
axes[2].set_title('Blood Pressure VS Age', fontsize=15, fontweight='bold')

# Line Plot Creation of Cholestrol VS Age in 4th plot
sns.lineplot(x='age', y='chol', data=data, color='b', ax=axes[3])
axes[3].set_title('Cholestrol VS Age', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 11))
axes = axes.flatten()

# Line Plot Creation of Person's ST Depression vs age in first subplot
sns.lineplot(x='age', y='oldpeak', data=data, color='g', ax=axes[0])
axes[0].set_title('ST Depression VS Age', fontsize=15, fontweight='bold')

# Bar Plot Creation of ST depression VS Heart Disease in second subplot
sns.barplot(x='sex1', y='oldpeak', data=data, hue='sex1', legend=False,  palette='twilight_r', ax=axes[1])
axes[1].set_title('ST depression VS Heart Disease', fontsize=15, fontweight='bold')

# Bar Plot Creation of Exercise With Angina VS Heart Disease in 3rd subplot
sns.barplot(x='Heart_Disease', y='exang', data=data, palette='viridis', hue='Heart_Disease', legend=False, ax=axes[2])
axes[2].set_title('Exercise With Angina VS Heart Disease', fontsize=15, fontweight='bold')

# Bar Plot Creation of Exercise With Angina VS Gender in 4th plot
sns.barplot(x='sex1', y='exang', data=data,  hue='sex1', legend=False, palette='binary_r', ax=axes[3])
axes[3].set_title('Exercise With Angina VS Gender', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Bar Plot Creation of Fasting Blood Sugar VS Gender using MatplotLib and Seaborn

sns.barplot(y='fbs', x='sex1', data=data)
plt.title(' Fasting Blood Sugar VS Gender', fontsize=17)
plt.xlabel('Sex', fontsize=15)
plt.ylabel('Fasting Blood Sugar', fontsize=15)
plt.show()
```

```python
# Check for non-numeric columns
non_numeric_cols = data.select_dtypes(include=['object']).columns

# Convert non-numeric columns to numeric if possible
for col in non_numeric_cols:
    try:
        data[col] = pd.to_numeric(data[col])
    except:
        # Drop columns that cannot be converted
        data.drop(columns=[col], inplace=True)

# Recalculate the correlation matrix and create the heatmap
corr_matrix = data.corr()

plt.figure(figsize=(16, 9))
sns.heatmap(corr_matrix, annot=True, linewidth=3)
plt.show()
```
